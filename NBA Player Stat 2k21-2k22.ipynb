## Web Scraping with pandas

### Method 1

You can modularize the URL and year string of text that may be useful for programmatic access

```python
#The blocks
year = "2022"
url = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'


#Combine the year + URL string
url_fix = url.format(year) #Using method formats the specified value(s) and insert them inside the string's placeholder
url_fix 
```

Let's begin to programmatically return a list of URL given a list of years (e.g 2016,2017,2018,2019,2020)

```python
years = [2016,2017,2018,2019,2020]
url = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
    
for year in years:
    url_new = url.format(year)
    print(url_new)
```

## Data Cleaning

If you open the URL link, the table header is presented multiple times in several rows

```python
df2022 = df[0]
```

```python
df2022[df2022.Age == 'Age']
```

```python
len(df2022[df2022.Age == 'Age']) 
```

```python
df = df2022.drop(df2022[df2022.Age == 'Age'].index) #Drop the multiple header
```
